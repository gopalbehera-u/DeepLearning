```python
import cv2
```

```python
face=cv2.CascadeClassifier(r"C:\Users\beher\OneDrive\Desktop\DeepLearning\Computer vision\harcar\haarcascade_frontalface_default.xml")
# loading pre-trained  face detector into face variable
```

```python
img=cv2.imread(r"C:\Users\beher\OneDrive\Desktop\DeepLearning\cnn\happy\nani20.jpg")
```

```python
img
# img stored a numpy array containing pixel values
```

```python
# Check if the image is loaded correctly
if img is None:
    print("Error: Image not found or cannot be loaded!")
    exit()  # Exit if image is not loaded
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# converting img color bgr to gray
```

```python
faces=face.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5)
# it search the image and finds all the faces
```

```python
faces
```

```python
if len(faces)==0:
    print("No face found")


else:
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2) # it create a rectangle aroung face

        cv2.imshow("face detection",img)
        cv2.waitKey(0)
        # (x,y) starting point 
        # (x+w , y+h) ending point
        # the thickness of the rectangle border 2 here.
    
cv2.destroyAllWindows()
```
