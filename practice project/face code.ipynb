```python
import cv2
```

```python
face=cv2.CascadeClassifier(r"C:\Users\beher\OneDrive\Desktop\DeepLearning\Computer vision\harcar\haarcascade_frontalface_default.xml")
```

```python
img=cv2.imread(r'C:\Users\beher\OneDrive\Desktop\DeepLearning\practice project\wallpaper2you_73050.jpg')
```

```python
img
```

```python
if img is None:
    print('img not found')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2BGRA)# convert bgr to gray
gray
```

```python
faces=face.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5)
```

```python
faces
```

```python
if len(faces)==0:
    print("face not found ")
else:
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        cv2.imshow('face detect',img)
        cv2.waitKey(0)
cv2.destroyAllWindows()
```
