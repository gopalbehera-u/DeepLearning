```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
dataset=pd.read_csv(r'C:\Users\beher\OneDrive\Desktop\DeepLearning\rnn\tmdb_5000_movies.csv')
dataset.head(3)
```

```python
dataset=dataset['original_title']
```

```python
dataset
```

```python
movie_name=dataset.to_list()
```

```python
movie_name
```

```python
tokenizer=tf.keras.preprocessing.text.Tokenizer()

tokenizer.fit_on_texts(movie_name)

seq=tokenizer.texts_to_sequences(movie_name)
```

```python
seq[:10]
```

```python
tokenizer.word_index
```

```python
x=[]
y=[]

total_words_dropped=0

for i in seq:
    if len(i) > 1:
        for index in range(1,len(i)):
            x.append(i[:index])
            y.append(i[index])
    else:
        total_words_dropped+=1
print(" total single words dropped : ",total_words_dropped)
```

```python
x[:10]
```

```python
y[:10]
```

```python
x=tf.keras.preprocessing.sequence.pad_sequences(x)
```

```python
x
```

```python
x.shape
```

```python
y=tf.keras.utils.to_categorical(y)
```

```python
y
```

```python
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
vocab_size
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 14),
    tf.keras.layers.LSTM(100, return_sequences=True),
    tf.keras.layers.LSTM(100),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(vocab_size, activation='softmax'),
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=250)
```

```python
model.save('nwp.h5')
```

```python
vocal_array=np.array(list(tokenizer.word_index.keys()))
```

```python
vocal_array
```

```python
def make_prediction(text, n_words):
    for i in range(n_words):
        text_tokenize = tokenizer.texts_to_sequences([text])
        text_padded = tf.keras.preprocessing.sequence.pad_sequences(text_tokenize, maxlen=14)
        prediction = np.squeeze(np.argmax(model.predict(text_padded), axis=-1))
        prediction = str(vocal_array[prediction - 1])
        print(vocal_array[np.argsort(model.predict(text_padded)) - 1].ravel()[:-3])
        text += " " + prediction
    return text
```

```python
make_prediction("cloudy",10)
```
