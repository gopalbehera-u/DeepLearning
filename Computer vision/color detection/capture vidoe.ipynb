```python
import cv2
import numpy as np
```

```python
cap=cv2.VideoCapture(0)
```

```python
while True:
    _,frame=cap.read()
    hvs_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    cv2.imshow("Frame",frame)
    key=cv2.waitKey(1)
    if key==ord('q') or  key==27:
        break

cap.release()
cv2.destroyAllWindows()
```
