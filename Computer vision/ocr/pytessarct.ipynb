```python
import cv2
import imutils
import pytesseract
import numpy as np
import gradio as gr

path=pytesseract.pytesseract.tesseract_cmd=r"C:\Users\beher\OneDrive\Desktop\DeepLearning\Computer vision\ocr"
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

def number_plate_ocr(img):

    # Gradio gives RGB image
    image = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

    image = imutils.resize(image, width=600)
    original = image.copy()

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray = cv2.bilateralFilter(gray, 11, 17, 17)

    edged = cv2.Canny(gray, 30, 200)

    contours, _ = cv2.findContours(
        edged.copy(),
        cv2.RETR_LIST,
        cv2.CHAIN_APPROX_SIMPLE
    )

    contours = sorted(contours, key=cv2.contourArea, reverse=True)[:30]

    screenCnt = None

    for c in contours:
        perimeter = cv2.arcLength(c, True)
        approx = cv2.approxPolyDP(c, 0.018 * perimeter, True)

        if len(approx) == 4:
            screenCnt = approx
            break

    plate_img = None
    plate_text = "Number plate not detected"

    if screenCnt is not None:

        cv2.drawContours(original, [screenCnt], -1, (0,255,0), 3)

        mask = np.zeros(gray.shape, np.uint8)
        cv2.drawContours(mask, [screenCnt], 0, 255, -1)

        x, y, w, h = cv2.boundingRect(screenCnt)
        plate_img = image[y:y+h, x:x+w]

        if plate_img.size != 0:
            plate_text = pytesseract.image_to_string(
                plate_img,
                config="--psm 7"
            ).strip()

            plate_img = cv2.cvtColor(plate_img, cv2.COLOR_BGR2RGB)

    output = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

    return output, plate_img, plate_text


demo = gr.Interface(
    fn=number_plate_ocr,
    inputs=gr.Image(type="numpy", label="Upload Car Image"),
    outputs=[
        gr.Image(label="Detected Plate"),
        gr.Image(label="Cropped Number Plate"),
        gr.Textbox(label="OCR Result")
    ],
    title="🚗 Number Plate OCR",
    description="Upload a vehicle image to detect and read the license plate."
)

demo.launch()
```
